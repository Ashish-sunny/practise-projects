```python
#separating the input and output variables
df_x=df_sonar.drop(columns=[60])
df_y=df_sonar[[60]]
print(df_x.shape)
print(df_y.shape)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
le.fit(df_y)
y=le.transform(df_y)
y
```

```python
df_x=np.array(df_x)
df_x
```

```python
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split
def maxf1_score(clf,df_x,y):
    maxf=0
    rs=0
    for r_state in range(42,100):
        x_train,x_test,y_train,y_test=train_test_split(df_x,y,random_state = r_state,test_size=0.20,stratify=y)
        clf.fit(x_train,y_train)
        y_pred=clf.predict(x_test)
        tmp=f1_score(y_test,y_pred)
        
        print('random state:',r_state,"and f1 score:",tmp)  
        
        if tmp>maxf:
            maxf=tmp
            rs=r_state
            
    print()
    print()
    print("maximum f1_score is at random state:",rs,"and it is:",maxf)
```

```python
df_models=pd.DataFrame(columns=['model','f1_score','std'])
models=list()
f1_scr=list()
std=list()
```

```python
from sklearn.linear_model import LogisticRegression
import warnings
warnings.filterwarnings('ignore')
log_reg=LogisticRegression()
maxf1_score(log_reg,df_x,y)
```

```python
#lets compute the mean f1 score and standard deviation using cross_val_score
from sklearn.model_selection import KFold , StratifiedKFold
f1_score_logistic=np.array([])
skf=StratifiedKFold(n_splits=5)
for train_index, test_index in skf.split(df_x,y):
    print("TRAIN:", train_index,"TEST:", test_index)
    x_train, x_test =df_x[train_index], df_x[test_index]
    y_train, y_test =y[train_index], y[test_index]
    log_reg.fit(x_train,y_train)
    y_pred=log_reg.predict(x_test)
    tmp=f1_score(y_test,y_pred)
    f1_score_logistic=np.append(f1_score_logistic,[tmp])
print()
print()
print()
print('mean f1 scorefor logistic after stratified cross validation :',np.mean(f1_score_logistic))
print('standard deviation for logistic from mean f1 score is:',np.std(f1_score_logistic))
print('array of f1 scores:',f1_score_logistic)
models.append('logistic')
f1_scr.append(np.mean(f1_score_logistic))
std.append(np.std(f1_score_logistic))
```

```python
#lets try decision tree classifier
from sklearn.tree import DecisionTreeClassifier
dtc= DecisionTreeClassifier()
maxf1_score(dtc,df_x,y)
```

```python
#lets compute the mean f1 score and standard deviation using cross_val_score
f1_score_dtc=np.array([])
skf = StratifiedKFold(n_splits=5)
for train_index, test_index in skf.split(df_x,y):
    #print('TRAIN:', train_index,"TEST",test_index)
    x_train, x_test = df_x[train_index],df_x[test_index]
    y_train, y_test = y[train_index],y[test_index]
    dtc.fit(x_train,y_train)
    y_pred=dtc.predict(x_test)
    tmp=f1_score(y_test,y_pred)
    f1_score_dtc=np.append(f1_score_dtc,[tmp])
    
print("mean f1 score for decision after stratified cross validation :",np.mean(f1_score_dtc))
print("standarfd deviation for decision tree from mean f1 score is:",np.std(f1_score_dtc))
print("array of f1 score :",f1_score_dtc)
models.append("decision tree")
f1_scr.append(np.mean(f1_score_dtc))
std.append(np.std(f1_score_dtc))
```

```python
#lets use KNN
#for KNN we need to know the best value of k using grid search
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
kc=KNeighborsClassifier()
neighbors={"n_neighbors":range(1,30)}
clf = GridSearchCV(kc, neighbors, cv=5,scoring="f1")
clf.fit(df_x,y)
clf.best_params_
```

```python
#knn classifier with n_neighbors=3
kc=KNeighborsClassifier(n_neighbors=3)
maxf1_score(kc,df_x,y)
```

```python
#lets compute the mean f1 score and standard deviation using cross val score
import numpy as np
from sklearn.model_selection import StratifiedKFold
f1_score_kc=np.array([])
skf=StratifiedKFold(n_splits=5)
for train_index, test_index in skf.split(df_x,y):
    #print("TRAIN:", train_index,"TEST:", test_index)
    x_train, x_test =df_x[train_index],df_x[test_index]
    y_train, y_test = y[train_index], y[test_index]
    kc.fit(x_train,y_train)
    y_pred=kc.predict(x_test)
    tmp=f1_score(y_test,y_pred)
    f1_score_kc=np.append(f1_score_kc,[tmp])
    
print("mean f1 score for KNN after stratified cross validation:",np.mean(f1_score_kc))
print("standard deviation for knn mean f1 score is :", np.std(f1_score_kc))
print("array of f1 scores :", f1_score_kc)
models.append("KNN")
f1_scr.append(np.mean(f1_score_kc))
std.append(np.std(f1_score_kc))
```

```python
#lets use SVM classifier

from sklearn.svm import SVC
svc=SVC()
parameters={"kernel":["linear","poly","rbf"],"C":[0.001,0.01,0.1,1,10]}
clf=GridSearchCV(svc, parameters, cv=5,scoring="f1")
clf.fit(df_x,y)
clf.best_params_
```

```python
svc=SVC(kernel="linear",C=10)
maxf1_score(svc,df_x,y)
```

```python
#lets compute the mean f1 score and standard deviation using cross_val_score
import numpy as np
from sklearn.model_selection import StratifiedKFold
f1_score_svc=np.array([])
skf =StratifiedKFold(n_splits=5)
for train_index, test_index in skf.split(df_x,y):
    #print("TRAIN", train_index , "TEST", test_index)
    x_train, x_test =df_x[train_index],df_x[test_index]
    y_train, y_test =y[train_index], y[test_index]
    svc.fit(x_train,y_train)
    y_pred =svc.predict(x_test)
    tmp=f1_score(y_test,y_pred)
    f1_score_svc=np.append(f1_score_svc,[tmp])
    
print("mean f1 score for SVM after stratified cross validation :",np.mean(f1_score_svc))
print("standard deviation for SVM from mean f1 score is :",np.std(f1_score_svc))
print("array of f1 scores :",f1_score_svc)
models.append("SVM")
f1_scr.append(np.mean(f1_score_svc))
std.append(np.std(f1_score_svc))
```

```python
df_models['model']= models
df_models['f1_score']= f1_scr
df_models['std']=std
df_models
```

```python
#from the above table w.r.t f1_score SVM is performing the best
#so we will choose that model
#max score is achievedat random state 91

x_train,x_test,y_train,y_test = train_test_split(df_x,y,random_state = 91, test_size=0.20,stratify=y)
svc.fit(x_train,y_train)
y_pred=svc.predict(x_test)
```

```python
#model performance
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
