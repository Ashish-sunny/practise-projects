# data is cleaned and preprocessed so far!!!

lets distribute our dataset into x and y variables according to the data dependency and target label column

```python
x=df_o.iloc[:,0:-1]
x.head()
```

```python
y=df_o.iloc[:,-1]
y.head()
```

```python
x.shape, y.shape
```

```python
#removing skewness from x
for i in x.columns:
    if x.skew().loc[i]>0.55:
        x[i]=np.log1p(x[i])
    if x.skew().loc[i]<-0.55:
        x[i]=np.square(x[i])
```

```python
x.skew()
```

```python
#standard scaling the x attribute for better nachine learning
from sklearn.preprocessing import StandardScaler
ss= StandardScaler()
x=ss.fit_transform(x)
```

```python
y.head()
```

```python
#splitting data into train test and split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state = 41,test_size=0.20)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
df_models=pd.DataFrame(columns=['model','f1_score'])
models=list()
f1_scr=list()
```

```python
from sklearn.metrics import f1_score
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
```

```python
f1_score(y_test,y_pred)
```

```python
#checking through DTC
from sklearn.tree import DecisionTreeClassifier
dtc= DecisionTreeClassifier()
dtc.fit(x_train,y_train)
y_pred=dtc.predict(x_test)
```

```python
f1_score(y_test,y_pred)
```
